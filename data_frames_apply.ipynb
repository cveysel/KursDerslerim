```python
def salary_category(salary):
    if salary < 50000:
        return "low"
    if 50000 <= salary < 75000 :
        return "medium"
    else:
        return "high"
```

```python
df["Salary"].apply(salary_category)
```

```python
df["salary_category"] = df["Salary"].apply(salary_category)
```

```python
df
```

```python
def adjust_perfom_score(row):
    if row["Experience"] > 10:
        return row["Performance_Score"]+1
    else:
        return row["Performance_Score"]
```

```python
df["Adjusted_score"] = df.apply(adjust_perfom_score, axis=1)
df
```

```python
def adjust_salary(row):
    if row["Experience"] > 20:
        return row["Salary"]+12500
    else:
        return row["Salary"]

df["Adjusted_salary"] = df.apply(adjust_salary, axis=1)
df[df["Experience"] > 20]
```
