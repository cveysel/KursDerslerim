```python
from cProfile import label

import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
age_list = [10,20,30,30,45,45,50,50,55]
weight_list = [20,70,75,50,80,42,68,98,45]
```

```python
plt.plot(age_list,weight_list,"g")
#plt.savefig("test.png") #kaydetmek
plt.title("Age vs Weight")
plt.xlabel("Age")
plt.ylabel("Weight")
plt.show()
```

array ile aynı sonuç

```python
np_age_list = np.array(age_list)
np_weight_list = np.array(weight_list)
```

```python
plt.plot(np_age_list,np_weight_list,"r")
plt.xlabel("Age")
plt.ylabel("Weight")
plt.title("Age vs Weight")
plt.show()
```

```python
numpy_arry1 = np.linspace(0,10,9)
```

```python
numpy_arry1
```

```python
numpy_arry2 = numpy_arry1 ** 3
numpy_arry2
```

```python
plt.plot(numpy_arry1,numpy_arry2)
plt.show()
```

```python
plt.plot(numpy_arry1,numpy_arry2,"y*-") # * yıldız şeklinde gösteriyor, - ise düz gösteriyor , ikisini de koyarsak ikisini de gösterir. -- kesikli çizgi gösterir.
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(numpy_arry1,numpy_arry2,"r*-")
plt.subplot(1,2,2)
plt.plot(numpy_arry2,numpy_arry1,"y*-")

#subplot yan yana tablo koyabilmek için
```

```python
my_figure = plt.figure()
figure_axes = my_figure.add_axes([0.1,0.1,0.8,0.8])
figure_axes.plot(numpy_arry1,numpy_arry2,"y")
figure_axes.set_xlabel("x axis")
figure_axes.set_ylabel("y axis")
figure_axes.set_title("Graph Title")
plt.show()
```

```python
new_fig = plt.figure(dpi=100)
new_axes = new_fig.add_axes([0.1,0.1,0.9,0.9])
new_axes.plot(numpy_arry1,numpy_arry1**2,label="numpy array **2",color="g")
new_axes.set_xlabel("x axis")
new_axes.plot(numpy_arry1,numpy_arry1**3,label="numpy array **3",color="y")
new_axes.legend(loc="best") # renk ayırma(sol üst)
plt.show()
```
