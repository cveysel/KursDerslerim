```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

median, quartile, min max

```python
data = np.array([5, 7, 8, 9, 13, 18, 54, 12, 45, 21, 75, 32, 59])
```

```python
plt.figure(figsize=(6,5))
sns.boxplot(y=data)
plt.title('Box Plot of the Data')
plt.grid(True)
plt.show()
```

```python
df = sns.load_dataset("titanic")
```

```python
df
```

```python
plt.figure(figsize=(15,5))

plt.subplot(1,2,1)
sns.boxplot(x="class",y="age",data=df)
plt.title("Class and Age Distribution")
plt.xlabel("Class")
plt.ylabel("Age")

plt.subplot(1,2,2)
sns.boxplot(x="class",y="fare",data=df)
plt.title("Class and Fare Distribution")
plt.xlabel("Class")
plt.ylabel("Fare")

plt.tight_layout()
plt.show()
```
