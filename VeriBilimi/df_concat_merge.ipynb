MERGE

```python
merge_df1 = pd.read_csv("7-merge_data1.csv")
merge_df2 = pd.read_csv("7-merge_data2.csv")
```

```python
merge_df1
```

```python
merge_df2
```

```python
df_merged_table_outer = pd.merge(merge_df1,merge_df2,on="Employee_ID",how="outer")
df_merged_table_outer
```

```python
df_merged_table_inner = pd.merge(merge_df1,merge_df2,on="Employee_ID",how="inner")
df_merged_table_inner
```

```python
df_merged_table_left = pd.merge(merge_df1,merge_df2,on="Employee_ID",how="left")
df_merged_table_left
```

```python
df_merged_table_right = pd.merge(merge_df1,merge_df2,on="Employee_ID",how="right")
df_merged_table_right
```
