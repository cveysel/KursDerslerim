```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("athlete_events.csv")
```

```python
data.describe()
```

```python
data.head()
```

```python
data.isnull()
```

```python
data.info()
```

```python
plt.scatter("Height", "Weight", data=data)
plt.xlabel("Height")
plt.ylabel("Weight")
plt.title("Athlete Events")
plt.show()
```

```python
sns.set_style("whitegrid")
sns.scatterplot(x="Height", y="Weight", data=data)
plt.xlabel("Height of Athlete")
plt.ylabel("Weight of Athlete")
plt.title("Athlete Events")
plt.show()
```

```python
sns.set_style("whitegrid")
sns.scatterplot(x="Height", y="Weight",hue="Sex", data=data)# hue Ã¶ne Ã§Ä±karmak istediÄŸimiz Ã¶zellik
plt.xlabel("Height of Athlete")
plt.ylabel("Weight of Athlete")
plt.title("Athlete Events")
plt.show()
```

```python
sns.set_style("whitegrid")
sns.scatterplot(x="Height", y="Weight",hue="Sex",style="Medal",size="Age", data=data)# hue Ã¶ne Ã§Ä±karmak istediÄŸimiz Ã¶zellik
plt.xlabel("Height of Athlete")
plt.ylabel("Weight of Athlete")
plt.title("Athlete Events")
plt.show()
```

```python
sns.set_style("whitegrid")
sns.lineplot(x="Height",y="Weight",hue="Sex",data=data)
plt.xlabel("Height of Athlete")
plt.ylabel("Weight of Athlete")
plt.title("Athlete Events")
plt.show()
```

```python
sns.set_style("whitegrid")
sns.displot(x="Height",hue="Sex",data=data,kind="kde")
plt.ylabel("Frequency")
plt.xlabel("Height of Athlete")
plt.title("Athletes Height Distribution")
plt.show()

# kind="kde"histogram yerine Ã§izgi olarak gÃ¶steriyor
```

```python
sns.barplot(x="Medal",y="Height",hue="Sex",data=data)
plt.title("Medals By Height Distribution")
plt.show()
```

```python
sns.catplot(x="Medal",y="Height",hue="Sex",col="Season",data=data)
plt.show()
```
