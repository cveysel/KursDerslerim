```python
new_fig,new_axes = plt.subplots()

new_axes.plot(data1,data1+1,color="r",linewidth=3, label="data1")
new_axes.plot(data1,data1+3,color="y",linewidth=2,label="data2")
new_axes.plot(data1,data1+5,color="g",linewidth=1,label="data3",linestyle="-.")

new_axes.plot(data1,data1+6,color="g",linewidth=1,label="data4",linestyle="--",marker="o",markersize=8,markerfacecolor="red")
new_axes.plot(data1,data1+7,color="g",linewidth=1,label="data5",linestyle="--",marker="+",markersize=8,markerfacecolor="blue")


new_axes.legend(loc="upper left")

plt.show()
```
