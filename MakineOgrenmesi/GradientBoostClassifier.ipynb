```python
X = df.drop("target", axis=1)
y = df["target"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=15)
```

```python
# bunu aralarnda ok yksek korelasyon olan stunlar bulmak iin kullanabiliriz
def corelation_for_dropping(df,threshold):
    corr = df.corr()
    columns_to_drop = set()
    for x in range(len(corr.columns)):
        for j in range(x):
            if abs((corr.iloc[x,j])) > threshold:
                columns_to_drop.add(corr.columns[x])
    return columns_to_drop
```

```python
corelation_for_dropping(X_train,0.85)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
from sklearn.model_selection import GridSearchCV
```

```python
gb = GradientBoostingClassifier()
gb.fit(X_train,y_train)
```

```python
y_pred = gb.predict(X_test)
print("confusion_matrix\n",confusion_matrix(y_test,y_pred))
print("classification_report\n",classification_report(y_test,y_pred))
print("accuracy_score\n",accuracy_score(y_test,y_pred))
```

```python
param = {
    "loss": ["log_loss", "exponential"],
    "learning_rate": [0.01,0.02,0.1],
    "n_estimators": [50,100,150,200],
    "max_depth": [3,4,5,6],
    "subsample": [0.7,0.8,0.9,1],
}
```

```python
grid = GridSearchCV(GradientBoostingClassifier(),param,cv=5,n_jobs=-1,verbose=1)
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
y_pred2 = grid.predict(X_test)
print("confusion_matrix\n",confusion_matrix(y_test,y_pred2))
print("classification_report\n",classification_report(y_test,y_pred2))
print("accuracy_score\n",accuracy_score(y_test,y_pred2))
```
