YukarÄ±dakiler Ã§Ä±karÄ±lacak outlier datalar. bÃ¶ylece daha iyi sonuÃ§ veren model geliÅtirilecek.

```python
df = df[((df["x"]<10))]
df = df[((df["y"]<20))]
df = df[((df["z"]<10) & (df["z"]>2))]
df = df[((df["depth"]<75) & (df["depth"]>45))]
df = df[((df["table"]<72) & (df["table"]>48))]
```

```python
X = df.drop("price",axis = 1)
y =df["price"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=15)
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
```

```python
df.info()
```

```python
for col in ["cut","color","clarity"]:
    X_train[col]= label_encoder.fit_transform(X_train[col])
    X_test[col]= label_encoder.transform(X_test[col])
```

```python
X_train["cut"].value_counts()
```

```python
X_test["cut"].value_counts()
```

```python
X_train.describe()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
```

```python
linear = LinearRegression()
linear.fit(X_train_scaled,y_train)
y_pred = linear.predict(X_test_scaled)
mae = mean_absolute_error(y_pred, y_test)
mse = mean_squared_error(y_pred, y_test)
r2 = r2_score(y_pred, y_test)
print("\nLinear Regression Model")
print("MAE: ",mae)
print("MSE: ",mse)
print("R2: ",r2)
plt.scatter(y_test, y_pred, color = "red")
```

```python
from sklearn.svm import SVR
```

```python
svr = SVR()
svr.fit(X_train_scaled,y_train)
y_pred = svr.predict(X_test_scaled)
mae = mean_absolute_error(y_test,y_pred)
mse = mean_squared_error(y_test,y_pred)
r2 = r2_score(y_test,y_pred)
print("\nLinear Regression Model")
print("MAE: ",mae)
print("MSE: ",mse)
print("R2: ",r2)
plt.scatter(y_test,y_pred)
```

```python
#hyperparameter tuning
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    "kernel": ["linear", "rbf"],
    "C":[0.1,1,10,100,1000],
    "gamma":["scale","auto"]
}
```

```python
grid = GridSearchCV(estimator= SVR(), param_grid= param_grid, n_jobs= -1, cv= 5, verbose= 3)
grid.fit(X_train_scaled,y_train)
# verbose iÅlem hakkÄ±nda bilgi veriyor
# cv cross validation
# n_jobs =-1 ise iÅlemci gÃ¼cÃ¼nÃ¼ tamamen burayay aktarÄ±yor
```

```python
y_pred = grid.predict(X_test_scaled)
```

```python
grid.best_params_
```

```python
mae = mean_absolute_error(y_test,y_pred)
mse = mean_squared_error(y_test,y_pred)
r2 = r2_score(y_test,y_pred)
print("\nLinear Regression Model")
print("MAE: ",mae)
print("MSE: ",mse)
print("R2: ",r2)
plt.scatter(y_test,y_pred)
```
