```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from mlflow import sklearn
%matplotlib inline
```

```python
df = pd.read_csv("6-bank_customers.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.columns
```

```python
df.describe()
```

```python
df.isna().sum()
```

```python
X = df.drop("subscribed", axis=1)
y = df["subscribed"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=15)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logistic = LogisticRegression()
logistic.fit(X_train, y_train)
```

```python
y_pred = logistic.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
score = accuracy_score(y_pred, y_test)
print("score= ",score)
print(classification_report(y_pred, y_test))
print(confusion_matrix(y_pred, y_test))
```

```python
#hyperparameter tuning
```

```python
model = LogisticRegression()
```

```python
penalty = ["l1","l2","elasticnet"]
c_values = [100, 10, 1, 0.1, 0.01] # cezalandırma katsayısı gibi bir ley
solver = ["newton_cg","lbfgs","liblinear","sag","saga","newton-cholesky"]
```

```python
params = dict(penalty=penalty, C=c_values, solver=solver)
```

```python
params
```

```python
#grid search cv
```

```python
from sklearn.model_selection import GridSearchCV, StratifiedKFold
```

```python
cv = StratifiedKFold()
```

```python
grid = GridSearchCV(estimator=model, param_grid=params, cv=cv,scoring="accuracy", n_jobs=-1)
```

```python
grid
```

```python
grid.fit(X_train, y_train)
```

```python
grid.best_params_ #yukarıdaki sözlükten denedi ve en uygun parametreleri çıkardı.
```

```python
grid.best_score_
```

```python
y_pred = grid.predict(X_test)
```

```python
score = accuracy_score(y_pred, y_test)
print("score= ",score)
print(classification_report(y_pred, y_test))
print(confusion_matrix(y_pred, y_test))
```

```python
y_pred = grid.predict(X_test)
```

```python
score = accuracy_score(y_pred, y_test)
print("score= ",score)
print(classification_report(y_pred, y_test))
print(confusion_matrix(y_pred, y_test))
```
